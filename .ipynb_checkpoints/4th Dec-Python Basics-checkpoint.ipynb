# Python Docstring
Notes written by the programmer for a particular module/class/function -- use function.__doc__ to access the doc string

Used for documentation

It is the first line written after function/class definition

```python
def docstring():
    '''This is a print function doc string'''
    print('Hello world')
```

```python
docstring.__doc__
```

```python
print(print.__doc__)
```

```python
import pickle, numpy
print(pickle.__doc__, numpy.__doc__)
```
